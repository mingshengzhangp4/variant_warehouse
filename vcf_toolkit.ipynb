```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from scidbpy import connect
from scidbpy.schema_utils import rechunk
```

```python
sdb = connect()
```

## Helper Routines

```python
VARIANT          = sdb.wrap_array("KG_VARIANT")
GENOTYPE         = sdb.wrap_array("KG_GENOTYPE")
VGENE            = sdb.wrap_array("GENE_36")
VARIANT_POS_MASK = sdb.wrap_array("KG_VARIANT_POSITION_MASK")
VCHROMOSOME      = sdb.wrap_array("KG_CHROMOSOME")
VARIANT_MULT_VAL = sdb.wrap_array("KG_VARIANT_MULT_VAL")
VSAMPLE          = sdb.wrap_array("KG_SAMPLE")
```

```python
def align_variants_to_genes(target_array_name='VARIANT_GENE_36', recompute=False):
    
    if not recompute and target_array_name in sdb.list_arrays():
        return sdb.wrap_array(target_array_name)
    
    sdb.remove(target_array_name)
    
    var_redim = (VARIANT.apply('pos_bucket', "int64(pos/1000000)")
                 .redimension("<pos:int64> [chrom_id=0:*,1,0, pos_bucket=0:*,1,0, variant_id=0:*,10000,0]")
                 .eval())
                             
    gene_redim = (VGENE.apply('chrom', 'string(chromosome_nbr)')
                  .index_lookup(VCHROMOSOME, 'chrom', 'chrom_id'))
                 
    gene_redim_1 = (gene_redim.apply('pos_bucket', "int64(genomic_start) / 1000000")
                    .redimension("<genomic_start:uint64, genomic_end:uint64> [chrom_id=0:*,1,0, pos_bucket=0:*,1,0, gene_id_36=0:*,40000,0]")
                    .eval())
  
    gene_redim_2 = (gene_redim.apply("pos_bucket", "int64(genomic_end) / 1000000")
                    .redimension("<genomic_start:uint64, genomic_end:uint64> [chrom_id=0:*,1,0, pos_bucket=0:*,1,0, gene_id_36=0:*,40000,0]")
                    .eval())

    mask1 = (sdb.merge(var_redim, gene_redim_1)
             .filter("pos>=genomic_start and pos<=genomic_end")
             .apply('mask', 'bool(true)')
             .redimension("<mask:bool>[variant_id =0:*,10000,0, gene_id_36=0:*,40000,0]")
             .eval())

    mask2 = (sdb.merge(var_redim, gene_redim_2)
             .filter("pos>=genomic_start and pos<=genomic_end")
             .apply('mask', 'bool(true)')
             .redimension("<mask:bool>[variant_id =0:*,10000,0, gene_id_36=0:*,40000,0]")
             .eval())
                 
    result = sdb.merge(mask1, mask2).eval().rename(target_array_name)                                   
    return result
```

```python
def generate_covar_matrix(min_frequency = 0.5, recompute=False):
    """
    Compute the covariance of samples across common variants.
    
    Creates and stores the COVAR_MATRIX and COVAR_SELECTED_VARIANTS arrays
    """
    if not recompute:
        arrays = sdb.list_arrays()
        if "COVAR_MATRIX" in arrays and "COVAR_SELECTED_VARIANTS" in arrays:
            return sdb.wrap_array("COVAR_MATRIX")
        
    sdb.remove('COVAR_MATRIX')
    sdb.remove('COVAR_SELECTED_VARIANTS')
    
    selected_variants = VARIANT_MULT_VAL[VARIANT_MULT_VAL['af'] > min_frequency]
    selected_variants = sdb.merge(VARIANT, selected_variants)
    selected_variants = selected_variants.redimension("<signature:string> [variant_id=0:*,10000,0]")
    
    num_variants = int(selected_variants.count()[0])
    num_samples = int(VSAMPLE.count()[0])
    
    print "Generating matrix using %i variants by %i samples" % (num_variants, num_samples)
    
    dense_selected_variants = sdb.unique(selected_variants)
    dense_selected_variants = rechunk(dense_selected_variants, chunk_size=10000).eval()
    dense_selected_variants.rename("COVAR_SELECTED_VARIANTS")
    

    matrix = GENOTYPE.apply('variant_present', "double(iif(gt='0|0', 0.0, 1.0))")
    matrix = (sdb.merge(matrix, selected_variants)
              .index_lookup(dense_selected_variants, 'signature', 'dense_variant_id')
              .redimension("<variant_present:double NULL> [dense_variant_id=0:%i,10000,0, sample_id=0:%i,313,0]" % 
                           (num_variants - 1, num_samples - 1))
              .eval()
              .rename('COVAR_MATRIX')
              )
    return matrix
```

## Covariance Demo

```python
matrix = generate_covar_matrix()
```

```python
downsampled = matrix.regrid(10).toarray()
plt.figure(figsize=(8, 8), dpi=200)
plt.imshow(downsampled, cmap='hot', aspect=.3)
plt.xlabel('Sample')
plt.ylabel('Variant')
```

### Covariance Matrix for Sample pairs

Takes another few minutes to compute

```python
centered = matrix - matrix.mean(0)
cv = sdb.dot(centered.T, centered) / centered.shape[0]
```

```python
sdb.remove('COVAR_RESULT')
cv.eval().rename('COVAR_RESULT')
```

```python
plt.figure(figsize=(10, 8), dpi=200)
plt.imshow(cv.regrid(10).toarray(), cmap='hot')
plt.colorbar()
plt.title("Sample Covariance Matrix")
```

The bands in this image represent collections of samples that are mutually-covariant. We think that these bands may correspond to different ethnic groups

#### Computing the first few principal components of the sample set 

```python
cv_pca, = sdb.svd(cv, return_U=True, return_S=False, return_VT=False)
cv_pca.eval().rename('COVAR_PCA')
```

```python
data = cv_pca[:, 0:3].toarray()
```

```python
# the first 2 dimensions of the PCA projection reveals clusters in the sample data
from sklearn.cluster import KMeans
cluster_ids = KMeans(k=2).fit(data).predict(data)

# the first 2 dimensions of the PCA projection reveals clusters in the sample data
for i, color in enumerate(['#e41a1c', '#377eb8']):
    hit = cluster_ids == i
    plt.plot(data[hit, 0], data[hit, 1], marker='o', linestyle='none', color=color, alpha=0.3)
```

## Sample Queries

```python
def all_instances_of_variant(variant_signature="21:10435894 C>T"):
    """ Select all instances of a given variant """
    
    # select a particular variant
    selected_variants = VARIANT[VARIANT['signature'] == variant_signature]
    
    # Select all genotypes with a variation in at least one chromosome
    result = sdb.merge(GENOTYPE.filter("gt <> '0|0'"), selected_variants)
    result = sdb.merge(VSAMPLE, result)
    
    result = result[['signature', 'sample_name', 'gt']]
    return result
```

```python
def all_variants_in_genes(gene_symbols = ('MRAP', 'SOD1')):
    """
    Select all variants observed in a given gene list
    """
    # create VARIANT_GENE_36 if needed
    align_variants_to_genes()
    
    VARIANT_GENE = sdb.wrap_array('VARIANT_GENE_36')
    selected_genes = VGENE.filter(' or '.join("gene_symbol = '%s'" % g for g in gene_symbols))
    selected_probes = selected_genes["gene_symbol", "genomic_start", "genomic_end"]
    result = sdb.merge(VARIANT_GENE, selected_genes)
    result = sdb.merge(VARIANT, result)
    result = result[['gene_symbol', 'signature']]
    return result
```

```python
def lookup_by_freq_pos(chrom='21', 
                       pos_start=11000000, 
                       pos_end=12000000, 
                       min_freq=0.9):
    """
    Lookup variants based on allele frequency and genomic coordinates
    """
    
    # between is an AFL call
    # SciDB-Py has some syntactic sugar:
    # array.between(a, b, ...) is equivalent to
    # sdb.afl.between(array, a, b, ...)    
    result = VARIANT_POS_MASK.between('null', pos_start, 'null', 'null', pos_end, 'null')
    
    result = sdb.merge(result, VCHROMOSOME.filter("chrom = '21'"))
    result = sdb.merge(VARIANT_MULT_VAL, result)
    result = result[result['af'] > min_freq]
    result = result.redimension("<af:double null> [variant_id=0:*,10000,0,order_nbr=0:*,5,0]")
    
    #A more thorough lookup: the variant may have many alternates, the allele frequency is different for each alternate.
    #Make sure we match the right genotype.
    #Alternatively, we could split the variants into unique alternates at load time - also a very reasonable approach.
    result['alternate_no'] = 'string(order_nbr + 1)'
    result = sdb.merge(VARIANT, result)
    result = result[['signature', 'alternate_no']]
    result = sdb.merge(GENOTYPE, result)
    result = result.filter("substr(gt, 2,1) = alternate_no or substr(gt, 0,1) = alternate_no")
    result = sdb.merge(VSAMPLE, result)
    
    return result    
    
```
